FEATURE ENGINEERING

```python
from sklearn.model_selection import train_test_split
X = df.drop('Sleep Disorder',axis = 1)
y = df['Sleep Disorder']
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=42,test_size = 0.2)
X_train.shape,X_test.shape
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
```

```python
transform = ColumnTransformer(transformers=[
    ('c1',OrdinalEncoder(categories = [['Software Engineer', 'Doctor', 'Sales Representative', 'Teacher',
       'Nurse', 'Engineer', 'Accountant', 'Scientist', 'Lawyer','Salesperson', 'Manager']]),['Occupation']),
    ('c2',OrdinalEncoder(categories =[['Normal','Normal Weight','Overweight','Obese']]),['BMI Category']),
    ('c3',OrdinalEncoder(categories =[['126/83', '125/80', '140/90', '120/80', '132/87', '130/86',
       '117/76', '118/76', '128/85', '131/86', '128/84', '115/75',
       '135/88', '129/84', '130/85', '115/78', '119/77', '121/79',
       '125/82', '135/90', '122/80', '142/92', '140/95', '139/91',
       '118/75']]),['Blood Pressure']),
    ('c4',OneHotEncoder(drop = 'first'),['Gender']),
    ('c5',StandardScaler(),[ 'Age', 'Sleep Duration', 'Quality of Sleep',
       'Physical Activity Level', 'Stress Level','Heart Rate', 'Daily Steps']),
],remainder = 'passthrough')
```

```python
df.columns
```

```python
df.head(1)
```

```python
X_train1 = transform.fit_transform(X_train)
X_test1 = transform.fit_transform(X_test)
```

```python
X_train1.shape
```

```python
le = LabelEncoder()
y_train1 = le.fit_transform(y_train)
y_test1 = le.fit_transform(y_test)
```

```python
y_train1 = pd.DataFrame(y_train1,columns = ['Sleep Disorder'])
```

```python
from imblearn.combine import SMOTETomek
smp = SMOTETomek()
X_train_samp,y_train_samp = smp.fit_resample(X_train1,y_train1)
```

```python
y_train_samp.value_counts()
```

## **NEURAL NETWORK**

```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense,Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
callback = EarlyStopping(
    monitor='val_loss',
    min_delta=0.0001,
    patience=10,
    verbose=0,
    mode='auto',
    baseline=None,
)
```

```python
model = Sequential()
```

```python
model.add(Dense(128,activation = 'relu',input_dim = 11))
model.add(Dense(128,activation = 'relu'))
model.add(Dropout(0.3))
model.add(Dense(128,activation = 'relu'))
model.add(Dropout(0.3))
model.add(Dense(3,activation = 'softmax'))
```

```python
model.summary()
```

```python
model.compile(optimizer = 'Adam',loss = 'sparse_categorical_crossentropy',metrics = 'accuracy')
```

```python
history = model.fit(X_train_samp,y_train_samp,epochs = 1000,callbacks = callback,validation_data=(X_test1,y_test1))
```

```python
model.get_weights()[0]
```

```python
mpl.plot(history.history['accuracy'])
mpl.plot(history.history['val_accuracy'])
```

```python
mpl.plot(history.history['loss'])
mpl.plot(history.history['val_loss'])
```
